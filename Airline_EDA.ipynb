# 1. Flight Traffic <a id="flight_traffic"></a>

Types of Delays:


**airline_delay**: The cause of the cancellation or delay was due to circumstances within the airline's control (e.g. maintenance or crew problems, aircraft cleaning, baggage loading, fueling, etc.).

**weather_delay**: Significant meteorological conditions (actual or forecasted) that, in the judgment of the carrier, delays or prevents the operation of a flight such as tornado, blizzard or hurricane.

**air_system_delay**: Delays and cancellations attributable to the national aviation system that refer to a broad set of conditions, such as non-extreme weather conditions, airport operations, heavy traffic volume, and air traffic control.

**aircraft_delay**: A previous flight with same aircraft arrived late, causing the present flight to depart late.

**security_delay**: Delays or cancellations caused by evacuation of a terminal or concourse, re-boarding of aircraft because of security breach, inoperative screening equipment and/or long lines in excess of 29 minutes at screening areas.

Based on this above explaination, delays or cancellations caused by the airline are the type of weather delays that could be reduced. So we would want to filter the delays that were caused by the airline.

Questions we are trying to answer:
1. What factors contributes the most to flight delay across different airlines?
2. Which month has the most delay in 2017?
3. Which state in America has the most flight cancellation?
4. What are the conditions that led to the cancellation in Airports?

```python
#let's combine the year, month and day into actual date.
flight_traffic['date']=pd.to_datetime(flight_traffic[["year", "month", "day"]])
```

```python
flight_traffic.isnull().sum()
```

```python
# 8053
# airline_delay_cancelled.shape
8053/59632
```

```python
#some of the actual_departure cells are missing, since it's only 13% of the data, we can drop it
flight_traffic=flight_traffic[flight_traffic.actual_departure.notnull()]
```

```python
airline_delay_cancelled=flight_traffic[(flight_traffic['cancelled']==1) | (flight_traffic['airline_delay']>0.0)]
```

```python
airline_delay_cancelled.isnull().sum()
```

```python
#for this 231 rows, we don't know the reason why it got delay or maybe late within 15mins which mean
#they were not delay, so we can drop them for now
airline_delay_cancelled[airline_delay_cancelled.airline_delay.isnull()].head()
```

```python
airline_delay_cancelled=airline_delay_cancelled[airline_delay_cancelled.airline_delay.notnull()]
```

```python
#totoal number of mins delay due to various types of delays across different airlines
airline_delay_cancelled.groupby('airline_id')['airline_delay','weather_delay','air_system_delay','security_delay','aircraft_delay'].sum().plot(kind='bar')
```

Airline delay can be associted with Aircraft delay because if airline delay can cause the next flight with same aircraft arrived late as well. From this above plot, we can see that some airline like AA, DL, EV, OO are the airlines that have more airline delays than the aircraft delay. We can take a look for their stock performance to see if the total minutes caused by airline delay will affect their stock performance.

```python
airline_delay_cancelled.head()
```

```python
#Let's see what are the most popular routes among these airlines
airline_delay_cancelled[['airline_id','origin_airport','destination_airport']].groupby(['origin_airport','destination_airport'])\
.count().sort_values(by=['airline_id'],ascending=False).head(10)
```
